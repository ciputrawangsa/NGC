```python
class Catalog:
    def __init__(self):
        self.data = []
    
    def addBook(self):
        title = input("Masukkan judul buku: ")
        author = input("Masukkan penulis buku: ")
        unid = int(input("Masukkan nomor seri buku: "))
        self.data.append({"Buku": title, "Penulis": author, "Uniqe identifier": unid})
        print(f"Buku {title} dengan pengarang {author} SN {unid} berhasil ditambahkan.\n")

    def displayBook(self):
        if len(self.data) == 0:
            print("Buku belum tersedia.\n")
        else:
            print(" no | Judul | Pengarang | Nomor Seri ")
            for i, item in enumerate(self.data, start=1):
                print(f"{i} | {item['Buku']} | {item['Penulis']} | {item["Uniqe identifier"]}")

    def searchBook(self):
        searchTerm = input("Cari buku berdasarkan Judul / Pengarang: ")
        found = False
        for book in self.data:
            if searchTerm in book["Buku"] or searchTerm in book["Penulis"]:
                found = True
                print(f"Buku {book['Buku']}, Penulis {book['Penulis']} tersedia.\n")
        if not found:
            print("Buku tidak tersedia.\n")

    def removeBook(self):
        bookDelTerm = input("Masukkan judul buku yang ingin dihapus: ")
        removed = False
        for delbook in self.data:
            removed = True
            if bookDelTerm in delbook["Buku"]:
                self.data.remove(delbook)
                print(f"Buku {delbook["Buku"]} sudah dihapus.\n")
        if not removed:
            print("Buku tidak ada di dalam list")
        
    def menu():
        library = Catalog()

        while True:
            try: 
                print("============================")
                print("   Sistem Katalog Buku      ")
                print("============================")

                print("Silahkan pilih menu dibawah ini: ")
                print("1. Tambah buku")
                print("2. Lihat daftar buku")
                print("3. Hapus buku")
                print("4. Cari buku")
                print("5. Keluar\n")
                userInput = int(input(f"Masukkan angka untuk memilih menu: "))
                if userInput == 1:
                    library.addBook()
                elif userInput == 2:
                    library.displayBook()
                elif userInput == 3:
                    library.removeBook()
                elif userInput == 4:
                    library.searchBook()
                elif userInput == 5:
                    False
                    break
                else:
                    print("Masukkan angka yang sesuai untuk memilih menu")
            except ValueError: 
                print("Cukup pakai angka untuk memilih menu")

    menu()


```
